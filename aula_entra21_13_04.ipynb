```python
numero = int(input("Digite um numero: "))
if numero % 2 != 0:
  print(f"Numeor {numero} eh impar")
else:
  print(f"Numeor {numero} eh par")
```

```python
a = int(input("Digite o primeiro valo: "))
b = int(input("Digite o segundo valor: "))

operacao = int(input("Escolha uma opcao: \n 1 - somar \n 2 - diminuir \n 3 - multiplicar"))

if operacao == 1:
  print(f"O resultado da soma eh: {a + b}")
elif operacao == 2:
  print(f"O resultado da subtracao eh: {a - b}")
elif operacao == 3:
  print(f"O resultado da multiplicacao eh: {a * b}")
else:
  print("Operacao invalida tente novamente")
```
